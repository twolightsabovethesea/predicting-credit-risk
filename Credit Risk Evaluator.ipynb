```python
#import dependencies
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data
```

```python
dum_train = pd.get_dummies(train_df)
dum_train
```

```python
# Convert categorical data to numeric and separate target feature for testing data
dum_test = pd.get_dummies(test_df)
dum_test.head()
```

```python
dum_test.head(50)
```

```python
#view the columns in the training data
dum_train.columns
```

```python
#view the columns in the testing data to check for differences
dum_test.columns
```

```python
#check the number of columns to confirm that there is only one difference - training data has 'debt_settlement_flag_Y' while test data does not
print(len(dum_train.columns))
print(len(dum_test.columns))
```

```python
# add missing dummy variables to testing set
```

### Since the column missing in the test data would not contribute to the training of the model in a meaningful way and the 'debt_settlement_flag_N' column provides the information (any row would be flagged Y or N so we can extrapolate from the N column), the 'debt_settlement_flag_Y' column can be dropped from the training data instead of adding a dummy column to the test data.

```python
#drop the 'debt_settlement_flag_Y' column from the training data to resolve the difference between training and test data
dum_train_drop = dum_train.drop('debt_settlement_flag_Y', axis=1)
```

```python
# separate the target from the features for the training data and drop columns from X that would contaminate the training data
X_train = dum_train_drop.drop(['loan_status_high_risk', 'loan_status_low_risk'], axis=1)
y_train = dum_train_drop['loan_status_high_risk']
```

```python
# separate the target from the features for the test data and drop columns from X that would contaminate the test data
X_test = dum_test.drop(['loan_status_high_risk', 'loan_status_low_risk'], axis=1)
y_test = dum_test['loan_status_high_risk']
```

### For this data, I believe that a Random Forest model will perform better than a Logistic Regression model.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
LR = LogisticRegression()
LR.fit(X_train, y_train)
print(f"Logistic Regression Testing Data Score: {LR.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
RF = RandomForestClassifier(random_state=1, n_estimators=500)
RF.fit(X_train, y_train)
print(f'Random Forest Testing Data Score: {RF.score(X_test, y_test)}')
```

### The Random Forest model did perform better than the Logistic Regression model. However, neither did well and improvements need to be made. Since the Random Forest model did better to begin with, I would choose to work on improving that one and abandon the Logistic Regression model at this stage. 

```python
# Scale the data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Now that the data has been scaled, the performance of the models may change. However, I think that the Random Forest model is still likely to outperform the Logistic Regression model.

```python
# Train the Logistic Regression model on the scaled data and print the model score
LR2 = LogisticRegression()
LR2.fit(X_train_scaled, y_train)
print(f"Logistic Regression Testing Data Score: {LR2.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score

RF2 = RandomForestClassifier(random_state=1, n_estimators=500)
RF2.fit(X_train_scaled, y_train)
print(f'Random Forest Testing Data Score: {RF2.score(X_test_scaled, y_test)}')
```

### Scaling the data did not improve the Random Forest model's score; in fact, it scored slightly lower on the scaled data. I was wrong in my prediction that the Random Forest model would still perform better, than the Logistic Regression model this time. The Logistic Regression model's score improved significantly, from about 0.525 to about 0.722. This would seem to indicate that scaling the data is very important when working with a Logistic Regression model and less so when working with a Random Forest model. The performance of both models is not particularly impressive and in order for either to be useful modifications would need to be made to improve them significantly.
